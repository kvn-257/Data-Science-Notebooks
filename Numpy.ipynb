# **Question**

Find the area of the region which enclosed within a unit square excluding the area of a unit quarter circle. Do the same for higher dimensions. Draw some inferences on the results.

```python
ndim = 2
```

```python
npoints = 1000000
```

```python
points = np.random.rand(npoints,ndim)
```

```python
dfo = np.zeros((npoints,1))
c = 0
```

```python
%%time
for i in range(npoints):
  for j in range(ndim):
    dfo[i] +=  points[i,j] ** 2
  dfo[i] = np.sqrt(dfo[i])
  if(dfo[i] > 1):
    c += 1
```

```python
print("fraction = ",c/npoints)
```

```python
%%time
print("fraction = ",np.sum(np.sqrt(np.sum(points * points, axis = 1)) > 1)/npoints)
```

```python
def area_ndim(npoints,ndim):
  points = np.random.rand(npoints,ndim)
  print("fraction = ",np.sum(np.sqrt(np.sum(points * points, axis = 1)) > 1)/npoints)
```

```python
for i in range(2,12):
  area_ndim(npoints,i)
```
